## 1. Introduction

   The way how money supply, inflation, and interest rates affect each other is a frequently discussed topic in Economics. The changes in the money supply are considered determinative of macroeconomics variables the interest rate and inflation rate are and vice versa. The purpose of this data analysis is to investigate the relationship between money supply, interest rate, and inflation rate in Turkey. 

   After the 2008 Financial Crisis, Central Bank of the Republic of Turkey as well as many countries, has implemented a policy of increasing the money supply. Following this information, I choose **2010:1-2020:12** period money supply, interest rate, and inflation rate monthly data is selected by myself to analyze the relationship between those parameters. The data is acquired from Central Bank of the Republic of Turkey Electronic Data Delivery System: https://evds2.tcmb.gov.tr

   We can sum up the question as **"Are money supply, interest rates, and inflation rate statistics related to eachother? If yes how?"**

## 2. Manipulation & Visualization

### 2.1 Data Manipulation

```r
#Reads data. Read first 133 rows since rest is appendix and explanation section
df <- read.xlsx("data.xlsx", sheet = 1,
  startRow = 1,
  colNames = TRUE,
  rowNames = FALSE,
  detectDates = TRUE,
  skipEmptyRows = TRUE,
  skipEmptyCols = TRUE,
  rows = c(1:133),
  cols = NULL,
  check.names = FALSE,
  sep.names = ".",
)

#Adds data.table class to data frame and checks
DT <- data.table(df)
print("Classes of the DT object")
class(DT)

#Sets column names 
setnames(DT, c("date","interest_rate","money_supply", "inflation") )

print("Tail of the DT object")
tail(DT)

#There are some NA values in the table. I will fill NA values with previous month's data
DT[, inflation := inflation[1], .(cumsum(!is.na(inflation)))]
DT[, money_supply := money_supply[1], .(cumsum(!is.na(money_supply)))]

#Formats date as date format. Day is needed in order to convert so I added 01 to every month as day.
DT[, date := as.Date(paste(date, "01", sep = "-")), ]

print("Tail of the formatted DT object")
tail(DT)
```

### 2.2 Data Visualization 

#### 2.2.1 Line Plots

```r
#Visual representation of the data as line plots

p1 <- ggplot(DT, aes(x=date, y=interest_rate, group = 1)) +
  geom_line(size = 1, colour = "royalblue4") +
  labs(title = "Consumer Interest Rate in Turkey 2010-2020", 
       x = "Date",
       y = "Interest Rate") +
  theme_minimal() +
  theme(plot.title = element_text(size=11))

p2 <- ggplot(DT, aes(x=date, y=inflation)) +
  geom_line(size = 1, colour = "royalblue4") +
  labs(title = "Inflation in Turkey 2010-2020", 
       x = "Date",
       y = "Inflation Rate") +
  theme_minimal() +
  theme(plot.title = element_text(size=11))


p3 <- ggplot(DT, aes(x=date, y=money_supply, group = 1)) +
  geom_line(size = 1, colour = "royalblue4") +
  labs(title = "Money Supply in Turkey 2010-2020", 
       x = "Date",
       y = "Money Supply") +
  theme_minimal() +
  theme(plot.title = element_text(size=12))


grid.arrange(p1,p2,p3, ncol=2)
```

By looking at the changes of the data over years visually following inference can be made.

 *-*Interest rate and inflation rate are following a very similar trend even the small fluctuations are visible on both graphs.

 *-*Money supply tends to increase constantly.

 *-*In 2018 all three graphs have a sharp increase followed by a sharp decrease in 2019. The reason behind this is probably the current government's unorthodox policies on the economy, especially on the interest. Later usual repo and interest policies were applied by the Central Bank of the Republic of Turkey. 
    
#### 2.2.2 Histograms


```r
#Visual representation of the data as histograms

#Iterates over years, creates a plot every for every year and stores it in plot_list variable, then displays the lots
plot_list <- vector(mode = "list", length = 11)
j=1
for(i in 2010:2020){
start <- as.Date(paste(i, "01-01", sep = "-"))

end <- as.Date(paste(i, "12-31", sep = "-"))

x <- DT[date>=start & date<=end]

    p <- ggplot(x, aes(x= interest_rate)) +
  geom_histogram( aes(y=..density..),colour = "lightblue4", alpha=0.6, fill = "lightblue1", bins= 10) +
  geom_density(alpha=.3, colour = "royalblue4", fill = "royalblue1") +
  labs(x= paste(i, "Interest Rate", sep = " "),
       y="Density") +
  theme_minimal() +
  theme(axis.text=element_text(size=8), axis.title=element_text(size=9,face="italic"))
    
    plot_list[[j]] <- p
    j <- j+1
    }


do.call('grid.arrange',c(plot_list, ncol = 3, top="Histograms of Interest Rates Over Years"))
```

The histogram graphs are drawn on a yearly basis from 2010 to 2020. Graphs do not show a pattern for any known distribution.

When I look at the graphics especially with a focus on x-axises, we can see that 2018, and 2019 differ from the others due to the high increase I mentioned after the 2.2.1 Line Plots section.

## 3. Correlation Analysis

### 3.1  Generating and Visualisation of Correlation Matrix

```r
# Creates correlation with ggcorr()
correlation_matrix <- cor(DT[,-c(1)])
correlation_matrix
```

```r
# Visualizes correlation with ggcorr()
correlation_plot_full <- ggcorrplot(correlation_matrix, method = "circle",
           show.legend = TRUE, lab = FALSE, outline.col = "royalblue4",
           title = "Correlation Matrix of money supply, inflation, and interest rates", colors = c("firebrick4","white","royalblue4"),
           legend.title = "Correlation") + theme(plot.title=element_text(hjust=0.5), 
                                                 axis.text.x = element_text(angle = 30, size=9, vjust =0.5, hjust=0.5))

# Visualizes correlation with ggcorr(). Using lower presents a plot easier to read
correlation_plot_lower <- ggcorrplot(correlation_matrix, method = "square", type = "lower",
           show.legend = TRUE, lab = TRUE, lab_col="white", lab_size=2, digits = 2,
           title = "Lower Correlation Matrix of money supply, inflation, and interest rates", colors = c("firebrick4","white","royalblue4"),
           legend.title = "Correlation") + theme(plot.title=element_text(hjust=0.5), 
                                                 axis.text.x = element_text(angle = 0, vjust =0.5, hjust=0.5))



grid.arrange(correlation_plot_full,correlation_plot_lower, ncol=1)
```

The correlation matrix and its visualizations suggest that there are positive correlations between our 3 possible pairs. The strength of the correlations are Inflaton-Interest Rate (\~0,86), Inflaton-Money Supply (\~0,62), Interest Rate-Money Supply (\~0,49) respectively. We can have more detail in a correlogram.

```r
# Creates correlogram with ggpairs
ggpairs(DT[,-c(1)])+ theme_minimal() + labs(title ="Correlogram of money supply, inflation, and interest rates")
```

In the correlogram spotting, the correlation is even easier. The correlogram also gives us information about the significance of the correlation. *** means our correlations also significant (p-value < 0.001). However, to access the actual results we need to run Pearson's product-moment correlation test.

```r
# Test corellations in order to find p-value. We can decide if correlation is significant or not depending on the value.
print("Pearson's product-moment correlation between MONEY SUPPLY and INFLATION")
cor.test(DT$money_supply, DT$inflation, method = "pearson")
print("Pearson's product-moment correlation between MONEY SUPPLY and INTEREST RATE")
cor.test(DT$money_supply, DT$interest_rate, method = "pearson")
print("Pearson's product-moment correlation between INTEREST RATE and INFLATION")

cor.test(DT$interest_rate, DT$inflation, method = "pearson")
```

p-values for every pair is very small. The very small p-value allows us to reject H0 which states that there is no correlation between given variables. All our correlations are significant.

r-values also appear (Same values have appeared on the correlation matrix before) at the bottom line of every test. r-value can take values between 1 and -1 where r=1 means a perfect positive correlation, r=0 no correlation, r=-1 perfect negative correlation.

All three values are greater than zero confirming there is a postive relationship between all pairs. Interest Rate and Inflation almost perfectly correlated Money Supply and Inflation strongly correlated. Money Supply and Interest Rate strongly correlated
